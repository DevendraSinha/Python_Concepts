```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go
```

# SCATTER-PLOT

# Data

```python
np.random.seed(42)
random_x=np.random.randint(1,101,100)
random_y=np.random.randint(1,101,100)
```

# Scatter plot

### Data 

```python
data=[go.Scatter(x=random_x,y=random_y,mode='markers',
                 marker=dict(size=12,
                             symbol='pentagon',
                             line={'width':2},
                             color='rgb(51,204,153)'
                             ))]

```

###Layout

```python
layout=go.Layout(title='Scatterplot',
                 xaxis=dict(title='MY FIRST X AXIS'),
                 yaxis=dict(title='MY FIRST Y AXIS'),
                 hovermode='closest'
                 )
```

###plot

```python
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig,filename='Scatter.html')
```

#LINE-CHART

```python
np.random.seed(56)
random_x=np.linspace(0,1,100)
random_y=np.random.randn(100)
```

```python
data=[go.Scatter(x=random_x,y=random_y,mode='lines',name='myline',
                 marker=dict(size=12,
                             symbol='pentagon',
                             line={'width':2},
                             color='rgb(51,204,153)'
                             ))]
```

```python
layout=go.Layout(title='Line-Chart',
                 xaxis=dict(title='MY FIRST X AXIS'),
                 yaxis=dict(title='MY FIRST Y AXIS'),
                 hovermode='closest'
                 )
```

```python
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig,filename='Line-Chart.html')
```
