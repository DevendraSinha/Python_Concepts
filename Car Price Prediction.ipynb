# Feature Engineering

# Analyze Columns Containing NULL Values

# Get Unique Values Information

```python
dataset.nunique()
```

```python
#Dropped Car_Id since it has no significance
dataset.drop('car_ID',axis=1,inplace=True)
```

```python
dataset['fueltype'].unique()
```

```python
#CleanUp CarName Column
dataset['CarName'] = dataset['CarName'].str.split(' ',expand=True)[0]
```

```python
dataset.columns
```

```python
dataset['CarName'] = dataset['CarName'].replace({'maxda': 'mazda', 'nissan': 'Nissan', 'porcshce': 'porsche', 'toyouta': 'toyota', 
                            'vokswagen': 'volkswagen', 'vw': 'volkswagen'})
```

# Find the Categorical Columns

```python
categorical_cols=dataset.select_dtypes(include=['object']).columns
```

#Encode Categorical Variables

```python
categorical_cols
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for cols in categorical_cols:
  dataset[cols]=le.fit_transform(dataset[cols])
```

```python
dataset.corr()
```

# Separate Dependent and Independent Variables

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

# Split the dataset into test and train

```python
from sklearn.model_selection import train_test_split
x = dataset.drop(['price'],axis=1)
y = dataset['price']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x.head(3)
```

# Scaling of the Features

```python
from sklearn.preprocessing import scale
cols=x.columns
X=pd.DataFrame(scale(x))
X.columns=cols
X
```

# Training the Multiple Linear Regression Model on Training set.

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```
