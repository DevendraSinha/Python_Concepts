```python
# Drop Rows having NULL Values
df.dropna(inplace=True)
```

```python
# Drop rows having  unaccounted returns or cancellations based on Quantity
df=df[df['Quantity'] > 0 ]
```

```python
# Remove items that appear not to cost anything
df=df[df['Price'] > 0 ]
```

```python
# drop returned items
df = df[~df['Invoice'].astype(str).str.contains("C",na=False)]
```

```python
df['Customer ID'] = df['Customer ID'].astype(str) # convert customerID to a string
#df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"]).dt.date # Only reatin date part of Invoice Date
```

```python
# Identify Outliers
def find_remove_outlier(df,val,col):
  Q1,Q3=np.percentile(val, [5, 95])
  IQR=Q3-Q1
  minval=Q1-1.5*IQR
  maxval=Q3+1.5*IQR
  df=df[(df[col]>minval) & (df[col]<maxval)]
  return df

```

```python
df=find_remove_outlier(df,df['Price'],'Price')
```

```python
df=find_remove_outlier(df,df['Quantity'],'Quantity')
```
