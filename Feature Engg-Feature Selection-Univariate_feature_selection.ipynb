```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import pandas_profiling as pp
from sklearn.datasets import load_boston
from sklearn.datasets import load_iris
from numpy import array 
#from pandas_profiling import ProfileReport
```

# Read Data

```python
dataset = pd.read_csv('/content/data.csv')
dataset_bkp=dataset
y =dataset['diagnosis'].map({'B':0,'M':1})# target variable
X= dataset.drop((['diagnosis','id']), axis=1)
```

```python
cor_matrix =dataset.corr().abs()
```

```python
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
```

```python
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.90)]
```

#Split the data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

#Run SelectKbest to select the 5 best features

```python
from sklearn.feature_selection import SelectKBest, chi2,f_regression 

X_5_best= SelectKBest(f_regression, k=5)
#print(X_5_best)
x_train = x_train.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
X_5_best.fit_transform(x_train, y_train)
print("-----------------")
#print(X_5_best.scores_)
print("------PVALES----------")
print(X_5_best.pvalues_)
filter = X_5_best.get_support()
#print(filter)
new_feat = [] 
for bool, feature in zip(filter, x_train.columns):
 if bool:
  new_feat.append(feature)
print('The best features are:{}'.format(new_feat)) # The list of your
```
