# Feature Engineering -Date Time

```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
data=pd.read_csv("/content/loan.csv",usecols=['date_issued','date_last_payment'])
```

```python
data.info()
```

```python
data['issue_dt']=pd.to_datetime(data['date_issued'])
data['last_payment_dt']=pd.to_datetime(data['date_last_payment'])
```

# Extract Date of the Month



```python
data['issue_date_week']=data['issue_dt'].dt.week
data['last_payment_week']=data['last_payment_dt'].dt.week
```

```python
data.head()
```

# Extract Month of the Date

```python
data['issue_month']=data['issue_dt'].dt.month
data['last_payment_month']=data['last_payment_dt'].dt.month
```

```python
data.head()
```

# Extract Quarter

```python
data['issue_quarter']=data['issue_dt'].dt.quarter
data['last_payment_quarter']=data['last_payment_dt'].dt.quarter
```

```python
data[['issue_dt','issue_quarter']].head()
```

# Extract Year

```python
data['issue_year']=data['issue_dt'].dt.year
data['last_payment_year']=data['last_payment_dt'].dt.year
```

```python
data[['issue_year','issue_dt']].head()
```

# Extract Day

```python
data['issue_day']=data['issue_dt'].dt.day
data['last_payment_day']=data['last_payment_dt'].dt.day
```

```python
data[['issue_day','issue_dt']].head()
```

# Extract day of the week

```python
data['issue_dayofweek']=data['issue_dt'].dt.dayofweek
data['last_payment_dayofweek']=data['last_payment_dt'].dt.dayofweek
```

# Extract DayName

```python
data['issue_dayofweekname']=data['issue_dt'].dt.day_name()
```

```python
data[['issue_dt','issue_dayofweekname']].head()
```

```python
data['issue_dt_is_weekend'] = np.where(data['issue_dayofweekname'].isin(['Sunday', 'Saturday']), 1,0)
```

```python
data[['issue_dt_is_weekend','issue_dayofweekname','issue_dt']].head()
```

#Extract Time elapsed between dates


```python
data['time_elpased']=data['issue_dt']-data['last_payment_dt']
```

```python
data[['time_elpased','issue_dt','last_payment_dt']].head()
```

# calculate number of months passed between 2 dates

```python
data['time_elpased']=(data['issue_dt']-data['last_payment_dt'])/ np.timedelta64(1, 'M')
```
