```python
import csv
import pandas as pd
```

```python
filename = "/content/test.csv.csv"
```

```python
# initializing the titles and rows list
fields = []
rows = []
err_rows=[]
```

```python
with open(filename, 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
    # extracting field names through first row
    fields = next(csvreader)
    for row in csvreader:
      new_lst=[item.split('|') for item in row]
      x=iter(new_lst[0])
      if len(new_lst[0])>13:
        err_rows.append(new_lst[0])
      else:
        rows.append(new_lst[0])
      
         
```
