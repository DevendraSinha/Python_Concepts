```python
def read_data():
 heart_dis_df=pd.read_csv(r"/content/heart_disrease.csv")
 return heart_dis_df
```

```python
def identify_variable_type(heart_dis_df):
    categorical_val1 = []
    continous_val1 = []
    for column in heart_dis_df.columns:
        #print(f"{column} : {heart_dis_df[column].unique()}")
        if heart_dis_df[column].nunique() <= 10:
            categorical_val1.append(column)
        else:
            continous_val1.append(column)
    return categorical_val1,continous_val1
```

```python
heart_dis_df=read_data()
```
