```python
from sklearn.impute import SimpleImputer 
from sklearn.compose import ColumnTransformer
from sklearn.pipeline  import Pipeline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_housing=pd.read_csv(r"/content/my_data/train.csv")
```

```python
df_housing.head()
```

```python
data=df_housing[['Age','Fare','Survived']]
```

```python
def impute_na(df,variable,mean_med):
    return df[variable].fillna(mean_med)
```

```python
mead=data['Age'].median()
mn=data['Age'].mean()
```

```python
data['Age_Median']=impute_na(data,'Age',mead)
data['Age_Mean']=impute_na(data,'Age',mn)
```

```python
 col = [
'OverallQual',
'TotalBsmtSF', 
#'IstFirSF', 
'GrLivArea', 
'WoodDeckSF',
'BsmtUnfSF', 
'LotFrontage', 
'MasVnrArea', 
'GarageYrBlt', 
'SalePrice'
]
```

```python

df=pd.read_csv(r"/content/my_data/train1.csv",usecols=col)
```

```python
df.columns
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
numeric_feature_mean=['LotFrontage']
numeric_feature_median=['MasVnrArea','GarageYrBlt']
```

```python
#Set Pipeline
```

```python
numeric_mean_imputer=Pipeline(steps=[('imputer',SimpleImputer(strategy='mean'))])
numeric_median_imputer=Pipeline(steps=[('imputer',SimpleImputer(strategy='median'))])
```

```python
# Pass the Pipeline to Column Transformer
```

```python
preprocessor=ColumnTransformer(transformers=[('mean_imputer',numeric_feature_mean,numeric_mean_imputer),
                                             ('median_imputer',numeric_feature_median,numeric_median_imputer)],
                               remainder='passthrough')
```

```python
preprocessor.fit(df)
```

```python
data=preprocessor.transform(data)
data.isnull.mean()
```
