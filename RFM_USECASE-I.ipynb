```python
df_data["revenue"]=(df_data["Unit Price"]-(df_data["Unit Price"]*df_data["Discount Applied"])- df_data["Unit Cost"])*df_data["Order Quantity"]
```

Data Type Conversion of Required column for RFM Analysis
The RFM Analysis is being Perfomed based on Customer Id
Recency -> Order Date (Total Order by the customer); Frequency --> Order Number   ; Monetary --> revenue

```python
df_data["_CustomerID"]=df_data["_CustomerID"].astype(str)
df_data["OrderDate"]=df_data["OrderDate"].apply(pd.to_datetime)  # Here I used apply method
```

**Calculate the number of days for Recency .Since the data set is very old we can use a more recent data using a variable**

```python


today_date=pd.to_datetime("2021-01-01")

# Get the data for caluculation of therespective score i.e
# for each customer  calculate
#   count the number of OrderNumber ,
#   sum of Revenue
#   number of days last purchase was done

df_rfm=df_data.groupby("_CustomerID").agg(
    {"OrderNumber": "count",
     "revenue":"sum",
      "OrderDate": lambda x: (today_date-x.max()).days
     }
 )
df_rfm.rename(columns={"OrderNumber":"Frequency","OrderDate":"Recency","revenue":"Monetary"},inplace=True)
df_rfm.to_csv("df_rfm1.csv")
df_rfm.head()

```
