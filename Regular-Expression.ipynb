# Metacharacters

### #1.\  (Backslash): we use the backslash(\) just before the special character so that it will lose its specialty

```python
base_str="geeks. for seek"
match=re.search(r'.',base_str)
print("Without ", match)
match=re.search(r'\.',base_str)
print("With ", match)
```

### #2. [] – Square Brackets:Square Brackets ([]) represent a character class consisting of a set of characters that we wish to match. For example, the character class [abc] will match any single a, b, or cWe can also specify a range of characters using – inside the square brackets. For example, 

[0, 3] is sample as [0123]
[a-c] is same as [abc]
