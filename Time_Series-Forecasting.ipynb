# Time Series Forecasting Technique- 1.Moving Average 

```python
df['Month'] = pd.to_datetime(df['datesold'], format='%Y-%m')
data=df[['Month','price']]
data = data.set_index('Month')
```

```python
data.head()
```

```python
data.plot(figsize=(20, 4))
plt.grid()
plt.legend(loc='best')
plt.title('Price Variation')
plt.show(block=False)
```
