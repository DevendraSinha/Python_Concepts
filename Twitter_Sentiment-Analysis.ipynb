```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from sklearn.model_selection import train_test_split
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sentiments = SentimentIntensityAnalyzer()
```

# Read Data

```python
tweet_df=pd.read_csv(r"/content/Sentiment.csv",error_bad_lines=False)
```

```python
tweet_df.head(2)
```

```python
tweet_df.columns
```

```python
tweet=tweet_df[['text','sentiment']]
```

```python
tweet.head(5)
```

```python
tweet = tweet[tweet.sentiment != "Neutral"]
tweet['text']=tweet['text'].apply(lambda x: x.lower())
tweet['text']=tweet['text'].apply(lambda x: re.sub('[^a-zA-Z0-9\s]','',x))
```

```python
tweet.head(5)
```

```python
tweet["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in tweet["text"]]
tweet["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in tweet["text"]]
tweet["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in tweet["text"]]
```

```python
x = sum(tweet["Positive"])
y = sum(tweet["Negative"])
z = sum(tweet["Neutral"])

def sentiment_score(a, b, c):
    if (a>b) and (a>c):
        print("Positive ðŸ˜Š ")
    elif (b>a) and (b>c):
        print("Negative ðŸ˜  ")
    else:
        print("Neutral ðŸ™‚ ")
sentiment_score(x, y, z)
```
